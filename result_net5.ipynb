```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
import pylab
from pylab import rcParams
from __future__ import print_function
```

```python
df_blocking_hop2 = pd.read_csv('../20181017_net5_sp1_hop2/blocked_205_b.csv', names = ('load', 'speed', 'Without defragmentation', 'Conventional', 'Proposed', 'Conventional with SSR-LNO', 'Proposed with SSRR-LNO') )
df_blocking_hop3 = pd.read_csv('../20181017_net5_sp1_hop3/blocked_205_b.csv', names = ('load', 'speed', 'Without defragmentation', 'Conventional', 'Proposed', 'Conventional with SSR-LNO', 'Proposed with SSRR-LNO') )
df_blocking_hop4 = pd.read_csv('../20181017_net5_sp1_hop4/blocked_205_b.csv', names = ('load', 'speed', 'Without defragmentation', 'Conventional', 'Proposed', 'Conventional with SSR-LNO', 'Proposed with SSRR-LNO') )
df_blocking_hop5 = pd.read_csv('../20181017_net5_sp1_hop5/blocked_205_b.csv', names = ('load', 'speed', 'Without defragmentation', 'Conventional', 'Proposed', 'Conventional with SSR-LNO', 'Proposed with SSRR-LNO') )
```

```python
for index, row in df_blocking_hop2.iterrows():
    df_blocking_hop2.loc[index, 'Without defragmentation':'Proposed with SSRR-LNO'] /= [10000, 10000, 10000, 10000, 10000]
for index, row in df_blocking_hop3.iterrows():
    df_blocking_hop3.loc[index, 'Without defragmentation':'Proposed with SSRR-LNO'] /= [10000, 10000, 10000, 10000, 10000]
for index, row in df_blocking_hop4.iterrows():
    df_blocking_hop4.loc[index, 'Without defragmentation':'Proposed with SSRR-LNO'] /= [10000, 10000, 10000, 10000, 10000]
for index, row in df_blocking_hop5.iterrows():
    df_blocking_hop5.loc[index, 'Without defragmentation':'Proposed with SSRR-LNO'] /= [10000, 10000, 10000, 10000, 10000]
```

```python
df_blocking_hop2 =  df_blocking_hop2.replace(0, np.NaN)
df_blocking_hop3 =  df_blocking_hop3.replace(0, np.NaN)
df_blocking_hop4 =  df_blocking_hop4.replace(0, np.NaN)
df_blocking_hop5 =  df_blocking_hop5.replace(0, np.NaN)
```

```python
plt.clf()
pylab.figure(figsize=(7, 7))
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.gca().yaxis.set_ticks_position('left')
plt.gca().xaxis.set_ticks_position('bottom')
plt.rcParams['font.family'] ='Times New Roman'
plt.rcParams['font.weight'] = 'light'
plt.rcParams['font.size'] = 22
plt.rcParams['axes.linewidth'] = 0.8
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 20
plt.rcParams['legend.borderaxespad'] = 0
plt.rcParams['legend.frameon'] = False
plt.rcParams['legend.numpoints'] = 1
plt.rcParams['legend.labelspacing'] = 0.1
plt.rcParams['savefig.bbox'] = 'tight'
```

```python
plt.yscale("log")
plt.minorticks_off()
plt.plot(df_blocking_hop2['load'], df_blocking_hop2['Without defragmentation'], marker='x',  color="black")
plt.plot(df_blocking_hop2['load'], df_blocking_hop2['Conventional'], label=('Conventional'), marker='^',  color="red",  linestyle='-')

plt.plot(df_blocking_hop2['load'], df_blocking_hop2['Proposed'], label=('Proposed (' + r'$\sigma$' + ' = 2)'), marker='.',  color="blue", linestyle='-')

plt.plot(df_blocking_hop3['load'], df_blocking_hop3['Proposed'], label=('Proposed (' + r'$\sigma$' + ' = 3)'), marker='.',  color="blue", linestyle='--')

plt.plot(df_blocking_hop4['load'], df_blocking_hop4['Proposed'], label=('Proposed (' + r'$\sigma$' + ' = 4)'), marker='.',  color="blue", linestyle='-.')

plt.plot(df_blocking_hop5['load'], df_blocking_hop5['Proposed'], label=('Proposed (' + r'$\sigma$' + ' = 5)'), marker='.',  color="blue", linestyle=':')

plt.yticks([1, 0.1, 0.01, 0.001])

plt.xlabel("Traffic load [Erlang]", weight=1) # x軸のタイトル
plt.ylabel("Blocking probability") # y軸
plt.legend(loc="lower right")
plt.savefig("../../sawa_paper/sawa/20180921_journal/fig/net5_hop.pdf")
plt.show()
plt.close()
```
